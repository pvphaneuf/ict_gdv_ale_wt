Using mutation convergence instead of frequency since there are no population samples within the IND CHEM TOL experiment.  
Using the mutation dataframe used to generate Fig 1 since it includes the same mutation filtering for comparable convergence results between this fig and Fig 1.

```python
import pandas as pd

all_AVA_muts = pd.read_pickle("./data/4_10_with_uniq_midpts.pkl")  # data from gd_ale_analytics repo
print(len(all_AVA_muts))
all_AVA_muts.head()
```

```python
TOL_EXP_L = [
    'TOL_2,3-butanediol', 'TOL_adipic_acid', 'TOL_coumaric_acid',
    'TOL_glutaric_acid', 'TOL_hexamethylenediamine',
    'TOL_hexanoic_acid', 'TOL_isobutyric_acid', 'TOL_n-butanol',
    'TOL_octanoic_acid', 'TOL_propanediol', 'TOL_putrescine']

tol_mut_df = all_AVA_muts[all_AVA_muts["exp"].isin(TOL_EXP_L)]
tol_mut_df
```

```python
supp_exp_ale_cnt_d = dict()
for exp, exp_mut_df in tol_mut_df.groupby(["supplement"]):
    supp_exp_ale_cnt_d[exp] = len(exp_mut_df["exp ALE"].unique())
supp_exp_ale_cnt_d
```

```python
TOL_PROJ_SEL_PRES = [
    '2 3-butanediol', 'adipic acid', 'coumaric acid', 'glutaric acid',
    'hexamethylenediamine', 'hexanoic acid', 'isobutyric acid',
    'n-butanol', 'octanoic acid', 'propanediol', 'putrescine']
```

```python
tol_chem_geno_feat_cnt_d = {x:dict() for x in TOL_PROJ_SEL_PRES}
for _, m in tol_mut_df.iterrows():
    for f in m["genomic features"]:
        if f["name"] not in tol_chem_geno_feat_cnt_d[m["supplement"]].keys():
            tol_chem_geno_feat_cnt_d[m["supplement"]][f["name"]] = {"ALE-unique mutation count":0, "log-odds":0, "exp ALEs":set()}
        tol_chem_geno_feat_cnt_d[m["supplement"]][f["name"]]["ALE-unique mutation count"] += 1
        tol_chem_geno_feat_cnt_d[m["supplement"]][f["name"]]["exp ALEs"].add(m["exp ALE"])
```

```python
tol_chem_geno_feat_df = pd.DataFrame()
for chem, geno_feat_mut_cnt_d in tol_chem_geno_feat_cnt_d.items():
    for geno_feat, d in geno_feat_mut_cnt_d.items():
        s = pd.Series({
            "chemical": chem,
            "geno feat": geno_feat,
            "ALE-unique mutation count": d["ALE-unique mutation count"],
            "exp ALE count": len(d["exp ALEs"])
            })
        tol_chem_geno_feat_df = pd.concat([tol_chem_geno_feat_df, pd.DataFrame(s).T], ignore_index=True)
tol_chem_geno_feat_df
```

```python
# tol_chem_geno_feat_df.set_index("geno feat", inplace=True)
tol_chem_geno_feat_df["ALE-unique mutation count"] = tol_chem_geno_feat_df["ALE-unique mutation count"].astype(int)
tol_chem_geno_feat_df["exp ALE count"] = tol_chem_geno_feat_df["exp ALE count"].astype(int)
tol_chem_geno_feat_df["ALE experiment convergence"] = tol_chem_geno_feat_df.apply(lambda r: r["exp ALE count"]/supp_exp_ale_cnt_d[r["chemical"]], axis=1)
tol_chem_geno_feat_df["geno feat"] = tol_chem_geno_feat_df["geno feat"].apply(lambda gf_name: gf_name.replace("attenuator terminator", "att term"))
tol_chem_geno_feat_df["geno feat"] = tol_chem_geno_feat_df["geno feat"].apply(lambda gf_name: gf_name.replace("terminator", "term"))
tol_chem_geno_feat_df
```

```python
rename_d = {
    '2 3-butanediol': 'butanediol',
    'adipic acid': 'adipate',
    'coumaric acid': 'coumarate',
    'glutaric acid': 'glutarate',
    'hexamethylenediamine': 'diaminohexane',
    'hexanoic acid': 'hexanoate',
    'isobutyric acid': 'isobutyrate',
    # 'n-butanol'
    'octanoic acid': 'octanoate',
    # 'propanediol'
    'putrescine': 'diaminobutane'
}
tol_chem_geno_feat_df['chemical'] = tol_chem_geno_feat_df['chemical'].apply(lambda c: rename_d[c] if c in rename_d.keys() else c)
tol_chem_geno_feat_df
```

```python
from util.const import convrg_cond_gene_d

all_genes_of_interest = set()
for v in convrg_cond_gene_d.values():
    all_genes_of_interest.update(v)
```

```python
all_cond_genes_of_interest_convrg_df = pd.DataFrame()
for c in convrg_cond_gene_d.keys():
    cdf = tol_chem_geno_feat_df[tol_chem_geno_feat_df.chemical==c]
    all_cond_genes_of_interest_convrg_df = pd.concat([all_cond_genes_of_interest_convrg_df, cdf[cdf["geno feat"].isin(all_genes_of_interest)]])

all_cond_genes_of_interest_convrg_df.head()
```

```python
all_cond_genes_of_interest_convrg_df.rename(columns={'geno feat': 'target', 'chemical': 'Condition1'}, inplace=True)
all_cond_genes_of_interest_convrg_df.head()
```

```python
# Get both key genes and DCA-significant genes for each condition for later plotting
# We don't want to plot all 22 genes per condition, only the ones that are key and/or DCA-significant.
# There's no guarantee that the key genes are DCA-significant, so we need to combine both sets.
# Want to keep those key genes without highly enriched variants per condition plot to emphasize being missed,
# so as to speak towards epistasis.
all_cond_genes_of_interest_d = dict()
for c in all_cond_genes_of_interest_convrg_df.Condition1.unique():
    cond_genes_of_interest = set(all_cond_genes_of_interest_convrg_df[all_cond_genes_of_interest_convrg_df.Condition1==c].target.unique())
    cond_signif_and_interest_genes = cond_genes_of_interest | set(all_cond_signif_hi_score_var_df[all_cond_signif_hi_score_var_df.condition==c].target.unique())
    all_cond_genes_of_interest_d[c] = cond_signif_and_interest_genes
```

```python
# Below is defined according to frequency of each
VAR_PRED_EFF_ORDER = [
    'moderately conservative',
    'conservative',
    'moderately radical',
    'radical',
    'truncation',
    'other',
    ]
```

```python
import seaborn as sns
import matplotlib.pyplot as mpl
sns.set_context("paper")
sns.set_style("whitegrid")
mpl.rcParams["figure.dpi"] = 150

import util.const as const


fig = mpl.figure(figsize=(10,8), layout="constrained")  # similar to fig.tight_layout()

row_count = 3
col_count = 3
subfigs = fig.subfigures(row_count, col_count)

# conditions ultimately bounded by where we had highly-enriched variants
conds = list(all_cond_signif_hi_score_var_df.condition.unique())

COND_KEY_GENE_COLOR = '#FECB52'

for iy in range(0, row_count):
    for ix in range(0, col_count):

# for iy in range(0, 1):
#     for ix in range(0, 1):

        ic = iy*3+ix
        if ic > len(conds)-1: break # break out of loop if we've reached the end of the conditions
        c = conds[ic]

        if ((c in all_cond_signif_hi_score_var_df.condition.unique())
            & (c in all_cond_genes_of_interest_convrg_df.Condition1.unique())):

            axs = subfigs[iy,ix].subplots(1, 3)

            # Getting genes all genes that host significant variants and/or have convergence values for condition.
            cond_convrg_and_or_signif_genes = all_cond_genes_of_interest_d[c]

            # Getting convergence values to plot
            cond_genes_of_interest_convrg_df = all_cond_genes_of_interest_convrg_df[
                (all_cond_genes_of_interest_convrg_df.Condition1 == c)
                & (all_cond_genes_of_interest_convrg_df.target.isin(cond_convrg_and_or_signif_genes))
                ][['target', 'ALE experiment convergence']]
            
            # Finding order of genes across plots according to convergence values.
            grouped = cond_genes_of_interest_convrg_df.groupby(["target"])
            df2 = pd.DataFrame({col:vals['ALE experiment convergence'] for col,vals in grouped})
            maxes = df2.max()
            maxes.sort_values(ascending=False, inplace=True)
            convrg_index_order = list(maxes.index)
            # for g in dca_cond_genes_of_interest_df.target.unique():
            for g in cond_convrg_and_or_signif_genes:
                if g not in convrg_index_order:
                    convrg_index_order.append(g)
            cond_genes_of_interest_convrg_df.set_index('target', inplace=True)
            cond_genes_of_interest_convrg_df = cond_genes_of_interest_convrg_df.reindex(
                index=convrg_index_order[::-1]  # I think this reversing is specific to pandas barplotting
                ).fillna(0)
            
            # Coloring key gene bars in convergence plot
            cond_key_genes = const.convrg_cond_gene_d[c]
            cond_genes_of_interest_convrg_df['color'] = cond_genes_of_interest_convrg_df.apply(lambda r: COND_KEY_GENE_COLOR if r.name in cond_key_genes else 'white', axis=1)
            
            cond_genes_of_interest_convrg_df['ALE experiment convergence'].plot(
                kind='barh', ax=axs[0], width=0.8, color=cond_genes_of_interest_convrg_df['color'], edgecolor='black')
            axs[0].set_xlim([0, 1.1])
            axs[0].set_ylabel("")

            dca_cond_genes_of_interest_df = all_cond_signif_hi_score_var_df[
                (all_cond_signif_hi_score_var_df.condition == c)
                & (all_cond_signif_hi_score_var_df.target.isin(cond_convrg_and_or_signif_genes))]

            sns.boxplot(
                ax=axs[1],
                data=dca_cond_genes_of_interest_df,
                x='log2FoldChange',
                y='target',
                order=convrg_index_order,
                flierprops = dict(marker='.', markeredgecolor='none',),
                boxprops=dict(facecolor='none'),
                width=0.8)
            sns.stripplot(
                ax=axs[1],
                data=dca_cond_genes_of_interest_df,
                x='log2FoldChange',
                y='target',
                hue='variant source',
                hue_order=['WT alleleome', 'ICT ALE', 'ALEdb'],  # To match order with barplot.
                order=convrg_index_order,
                size=3,
                palette = const.VAR_SRC_COLOR_D,
                alpha=1,
                dodge=True)
            axs[1].get_legend().remove()
            axs[1].yaxis.set_ticklabels([])
            axs[1].set_ylabel("")
            axs[1].yaxis.grid(True)
            axs[1].set_xlabel("")

            sns.boxplot(
                ax=axs[2],
                data=dca_cond_genes_of_interest_df,
                x='log2FoldChange',
                y='target',
                order=convrg_index_order,
                flierprops = dict(marker='.', markeredgecolor='none',),
                boxprops=dict(facecolor='none'))
            sns.stripplot(
                ax=axs[2],
                data=dca_cond_genes_of_interest_df,
                x='log2FoldChange',
                y='target',
                hue='predicted variant effect',
                palette = var_eff_pal,
                order=convrg_index_order,
                hue_order=VAR_PRED_EFF_ORDER,
                size=3,
                alpha=1,
                dodge=True)
            axs[2].get_legend().remove()
            axs[2].yaxis.set_ticklabels([])
            axs[2].set_ylabel("")
            axs[2].yaxis.grid(True)
            axs[2].set_xlabel("")

            subfigs[iy,ix].suptitle(c)

fig.supylabel('All key genes')
fig.supxlabel('ALE experiment convergence (left) and log2FoldChange colored by source (middle)\nand predicted variant effect (left)')
# fig.suptitle('Comparing ALE experiment key genes to significantly enriched Onyx designs')


from matplotlib.patches import Patch
custom_patches = [
    Patch(facecolor='#FECB52', edgecolor='black'),
    Patch(facecolor='white', edgecolor='black')]
subfigs[iy,ix].legend(custom_patches, ['Condition specific key gene', 'General gene of interest'], loc='upper left', bbox_to_anchor=(0.1, 0.95))

# from matplotlib.lines import Line2D
# legend1_handles = [Line2D([0], [0], marker='o', color='w', label=var_src, markerfacecolor=color, markersize=10) for var_src, color in const.VAR_SRC_COLOR_D.items()]
# subfigs[iy,ix].legend(handles=legend1_handles, title='Variant Sources', loc='upper left', bbox_to_anchor=(0.1, 0.95))
# legend2_handles = [Line2D([0], [0], marker='o', color='w', label=var_eff, markerfacecolor=color, markersize=10) for var_eff, color in var_eff_pal.items()]
# subfigs[iy,ix].legend(handles=legend2_handles, title='Variant Effects', loc='upper left', bbox_to_anchor=(0.1, 0.75))

fig.text(0, 1, 'C', weight='bold', fontsize=14)

mpl.savefig("f2c.svg")
```
