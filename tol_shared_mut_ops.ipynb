```python
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
all_AVA_muts = pd.read_pickle("./data/4_10_with_uniq_midpts.pkl")
print(len(all_AVA_muts))
all_AVA_muts.head()
```

```python
TOL_EXP_L = ['TOL_2,3-butanediol', 'TOL_adipic_acid', 'TOL_coumaric_acid',
       'TOL_glutaric_acid', 'TOL_hexamethylenediamine',
       'TOL_hexanoic_acid', 'TOL_isobutyric_acid', 'TOL_n-butanol',
       'TOL_octanoic_acid', 'TOL_propanediol', 'TOL_putrescine']
```

```python
tol_mut_df = all_AVA_muts[all_AVA_muts["exp"].isin(TOL_EXP_L)]
tol_mut_df
```

```python
feat_regdbid_to_name_d = dict()
for _, m in tol_mut_df.iterrows():
    for f in m["genomic features"]:
        feat_regdbid_to_name_d[f["RegulonDB ID"]] = f["name"]
    for f in m["operons"]:
        feat_regdbid_to_name_d[f["RegulonDB ID"]] = f["name"] + " operon"
```

```python
feat_regdbid_to_type_d = dict()
for _, m in tol_mut_df.iterrows():
    for f in m["genomic features"]:
        feat_regdbid_to_type_d[f["RegulonDB ID"]] = f["feature type"]
    for f in m["operons"]:
        feat_regdbid_to_type_d[f["RegulonDB ID"]] = "operon"
```

```python
geno_feat_set = set()
for _, m in tol_mut_df.iterrows():
    for geno_feat in m["genomic features"]:
        if geno_feat["RegulonDB ID"] != None:
            geno_feat_set.add(geno_feat["RegulonDB ID"])
```

```python
op_set = set()
for _, m in tol_mut_df.iterrows():
    for op in m["operons"]:
        if op["RegulonDB ID"] != None:
            op_set.add(op["RegulonDB ID"])
feat_set = op_set
```

```python
feat_set = op_set | geno_feat_set
```

```python
exp_feat_mut_cnt_mat = pd.DataFrame(columns=TOL_EXP_L, index=feat_set)
exp_feat_mut_cnt_mat = exp_feat_mut_cnt_mat.fillna(0)

for tol_exp, exp_mut_df in tol_mut_df.groupby(["exp"]):
    for _, m in exp_mut_df.iterrows():
        for f in m["genomic features"]:
            exp_feat_mut_cnt_mat.at[f['RegulonDB ID'], tol_exp] += 1
        for f, links in m["operon links"].items():
            exp_feat_mut_cnt_mat.at[f, tol_exp] += len(links)
            
def _get_feat_name(regdbid):
    feat_name = ''
    try:
        feat_name = feat_regdbid_to_name_d[regdbid]
    except:
        feat_name = regdbid
    return feat_name

exp_feat_mut_cnt_mat["observed experiment count"] = exp_feat_mut_cnt_mat.apply(lambda r: r.astype(bool).sum(), axis=1)

exp_feat_mut_cnt_mat["feature type"] = exp_feat_mut_cnt_mat.apply(lambda r: feat_regdbid_to_type_d[r.name], axis=1)
exp_feat_mut_cnt_mat["feature type"] = exp_feat_mut_cnt_mat["feature type"].apply(lambda x: "intergenic" if x == "unknown" else x)

exp_feat_mut_cnt_mat['ALE-unique mutation count'] = exp_feat_mut_cnt_mat.drop(['observed experiment count', 'feature type'], axis=1).sum(axis=1)
exp_feat_mut_cnt_mat.sort_values(by=["ALE-unique mutation count"], inplace=True, ascending=False)

exp_feat_mut_cnt_mat["name"] = exp_feat_mut_cnt_mat.apply(lambda r: _get_feat_name(r.name), axis=1)
exp_feat_mut_cnt_mat.set_index('name', inplace=True)

exp_feat_mut_cnt_mat
```

Find systems that are mutated across more than one experiment

```python
p = {
    'gene':"#F65E54",
    'pathway':"#72C4B3",
    'operon':"#A7A0CB",
    'regulon':"#BAE4B0",
    'TFBS':"#FA9A47",
    'attenuator terminator':"#9BD44C",
    'intergenic':"#F9B8DA",
    'promoter':"#CAC9CA",
    'terminator':"#A35DA6",
    'RBS':"#FFE953",
}
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from adjustText import adjust_text
%matplotlib inline
plt.rcParams["figure.dpi"] = 100
sns.set_context("paper")
sns.set_style("whitegrid")


fig, ax = plt.subplots(figsize=(10,4))


X_AXIS = "observed experiment count"
Y_AXIS = "ALE-unique mutation count"

ax = sns.scatterplot(
    ax=ax,
    data=exp_feat_mut_cnt_mat,
    x=X_AXIS,
    y=Y_AXIS,
    hue="feature type",
    linewidth=0,
    palette=p,
    legend="full",
)

annot_df = exp_feat_mut_cnt_mat[
    exp_feat_mut_cnt_mat[Y_AXIS]>=15
]

# To get the frequently mutated dots plotted on the foreground
sns.scatterplot(
    ax=ax,
    data=annot_df,
    x=X_AXIS,
    y=Y_AXIS,
    hue="feature type",
    linewidth=0,
    palette=p,
    legend=False
)

# ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.legend(loc='lower center', ncol=4, bbox_to_anchor=(0.5, 1))


texts = []
for x, y, s in zip(annot_df[X_AXIS].tolist(),
                   annot_df[Y_AXIS].tolist(),
                   annot_df.index.tolist()):
    texts.append(plt.text(x, y, s))
force = 0.6
adjust_text(texts,
            arrowprops=dict(arrowstyle="->", color='black', lw=0.5),
            force_points=((force, force)),
            force_text=((force, force))
           )

ax.text(0.1, 55, 'B', weight='bold', fontsize=14)
plt.savefig("f1b.svg")
```
