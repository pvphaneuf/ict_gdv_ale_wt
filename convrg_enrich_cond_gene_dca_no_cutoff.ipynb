# Get convergence values for genes of interest

```python
# Using previous AVA (or more likely gd_ale_analytics repo) published results because they're cleaned.

import pandas as pd

all_AVA_muts = pd.read_pickle("./data/4_10_with_uniq_midpts.pkl")
print(len(all_AVA_muts))
all_AVA_muts.head()
```

```python
TOL_EXP_L = [
    'TOL_2,3-butanediol', 'TOL_adipic_acid', 'TOL_coumaric_acid',
    'TOL_glutaric_acid', 'TOL_hexamethylenediamine',
    'TOL_hexanoic_acid', 'TOL_isobutyric_acid', 'TOL_n-butanol',
    'TOL_octanoic_acid', 'TOL_propanediol', 'TOL_putrescine']

tol_mut_df = all_AVA_muts[all_AVA_muts["exp"].isin(TOL_EXP_L)]
tol_mut_df
```

```python
supp_exp_ale_cnt_d = dict()
for exp, exp_mut_df in tol_mut_df.groupby(["supplement"]):
    supp_exp_ale_cnt_d[exp] = len(exp_mut_df["exp ALE"].unique())
supp_exp_ale_cnt_d
```

```python
TOL_PROJ_SEL_PRES = [
    '2 3-butanediol', 'adipic acid', 'coumaric acid', 'glutaric acid',
    'hexamethylenediamine', 'hexanoic acid', 'isobutyric acid',
    'n-butanol', 'octanoic acid', 'propanediol', 'putrescine']
```

```python
# Might be more informative to use "convergence" (ALE%) rather than "ALE-unique mutation count"
tol_chem_geno_feat_cnt_d = {x:dict() for x in TOL_PROJ_SEL_PRES}
for _, m in tol_mut_df.iterrows():
    for f in m["genomic features"]:
        if f["name"] not in tol_chem_geno_feat_cnt_d[m["supplement"]].keys():
            tol_chem_geno_feat_cnt_d[m["supplement"]][f["name"]] = {"ALE-unique mutation count":0, "log-odds":0, "exp ALEs":set()}
        tol_chem_geno_feat_cnt_d[m["supplement"]][f["name"]]["ALE-unique mutation count"] += 1
        tol_chem_geno_feat_cnt_d[m["supplement"]][f["name"]]["exp ALEs"].add(m["exp ALE"])

        # try:
        #     tol_chem_geno_feat_cnt_d[m["supplement"]][f["name"]]["log-odds"] = assoc_mat.at[f["name"], m["supplement"]]
        # except:
        #     print("no assoc data for :" + f["name"])  # Should figure out why this is the case
```

```python
tol_chem_geno_feat_df = pd.DataFrame()
for chem, geno_feat_mut_cnt_d in tol_chem_geno_feat_cnt_d.items():
    for geno_feat, d in geno_feat_mut_cnt_d.items():
        s = pd.Series({
            "chemical": chem,
            "geno feat": geno_feat,
            "ALE-unique mutation count": d["ALE-unique mutation count"],
            # "association to chemical (log-odds)": d["log-odds"],
            "exp ALE count": len(d["exp ALEs"])
            })
        tol_chem_geno_feat_df = pd.concat([tol_chem_geno_feat_df, pd.DataFrame(s).T], ignore_index=True)
tol_chem_geno_feat_df
```

```python
# tol_chem_geno_feat_df.set_index("geno feat", inplace=True)
tol_chem_geno_feat_df["ALE-unique mutation count"] = tol_chem_geno_feat_df["ALE-unique mutation count"].astype(int)
tol_chem_geno_feat_df["exp ALE count"] = tol_chem_geno_feat_df["exp ALE count"].astype(int)
tol_chem_geno_feat_df["ALE experiment convergence"] = tol_chem_geno_feat_df.apply(lambda r: r["exp ALE count"]/supp_exp_ale_cnt_d[r["chemical"]], axis=1)
tol_chem_geno_feat_df["geno feat"] = tol_chem_geno_feat_df["geno feat"].apply(lambda gf_name: gf_name.replace("attenuator terminator", "att term"))
tol_chem_geno_feat_df["geno feat"] = tol_chem_geno_feat_df["geno feat"].apply(lambda gf_name: gf_name.replace("terminator", "term"))
tol_chem_geno_feat_df
```

```python
rename_d = {
    '2 3-butanediol': 'butanediol',
    'adipic acid': 'adipate',
    'coumaric acid': 'coumarate',
    'glutaric acid': 'glutarate',
    'hexamethylenediamine': 'diaminohexane',
    'hexanoic acid': 'hexanoate',
    'isobutyric acid': 'isobutyrate',
    # 'n-butanol'
    'octanoic acid': 'octanoate',
    # 'propanediol'
    'putrescine': 'diaminobutane'
}
tol_chem_geno_feat_df['chemical'] = tol_chem_geno_feat_df['chemical'].apply(lambda c: rename_d[c] if c in rename_d.keys() else c)
tol_chem_geno_feat_df
```

```python
from util.const import convrg_cond_gene_d

key_genes = set()
for v in convrg_cond_gene_d.values():
    key_genes.update(v)
# key_genes
```

```python
key_genes_convrg_df = pd.DataFrame()
for cond in convrg_cond_gene_d.keys():
    cdf = tol_chem_geno_feat_df[tol_chem_geno_feat_df.chemical==cond]
    key_genes_convrg_df = pd.concat([key_genes_convrg_df, cdf[cdf["geno feat"].isin(key_genes)]])

key_genes_convrg_df.head()
```

```python
import util.const as const

var_eff_pal = const.GRANTHAM_CATEGORY_COLOR_PALETTE.copy()
var_eff_pal.update(const.PREDICTED_VARIANT_EFFECT_COLOR_PALETTE)
print(var_eff_pal)
```

```python
# Below is defined according to frequency of each
VAR_PRED_EFF_ORDER = [
    'moderately conservative',
    'conservative',
    'moderately radical',
    'radical',
    'truncation',
    'other',
    ]
```

```python
import seaborn as sns
import matplotlib.pyplot as mpl
mpl.rcParams["figure.dpi"] = 150
sns.set_context("paper")
sns.set_style("whitegrid")

import util.const as const


fig = mpl.figure(figsize=(10,8), layout="constrained")  # similar to fig.tight_layout()

row_count = 3
col_count = 3
subfigs = fig.subfigures(row_count, col_count)

# conditions ultimately bounded by what we have enrichment data for.
conds = list(dca_signif_df.condition.unique())

COND_KEY_GENE_COLOR = '#FECB52'

for iy in range(0, row_count):
    for ix in range(0, col_count):
# for iy in range(0, 1):
#     for ix in range(0, 1):
        
        ic = iy*3+ix
        if ic > len(conds)-1:
            break # break out of loop if we've reached the end of the conditions
        c = conds[ic]

        if (c in dca_signif_df.condition.unique()) & (c in key_genes_convrg_df.Condition1.unique()):

            axs = subfigs[iy,ix].subplots(1, 3)

            # all_cond_genes_of_interest = cond_genes_of_interest_d[c]

            # Getting regular DCA data since some key genes won't have been considered enriched
            dva_cond_df = dca_df[dca_df.condition == c]
            # dva_cond_genes_of_interest_df = dva_cond_df[dva_cond_df.target.isin(all_cond_genes_of_interest)]
            dva_cond_genes_of_interest_df = dva_cond_df[dva_cond_df.target.isin(key_genes)]

            convrg_cond_df = tol_chem_geno_feat_df[tol_chem_geno_feat_df.chemical == c]
            # convrg_cond_genes_df = convrg_cond_df[convrg_cond_df['geno feat'].isin(all_cond_genes_of_interest)]
            convrg_cond_genes_df = convrg_cond_df[convrg_cond_df['geno feat'].isin(key_genes)]
            convrg_cond_genes_df = convrg_cond_genes_df[['geno feat', 'ALE experiment convergence']]
            
            # For finding order of genes across plots according to convergence values.
            grouped = convrg_cond_genes_df.groupby(["geno feat"])
            df2 = pd.DataFrame({col:vals['ALE experiment convergence'] for col,vals in grouped})
            maxes = df2.max()
            maxes.sort_values(ascending=False, inplace=True)
            convrg_index_order = list(maxes.index)
            
            for g in dva_cond_genes_of_interest_df.target.unique():
                if g not in convrg_index_order:
                    convrg_index_order.append(g)

            convrg_cond_genes_df.set_index('geno feat', inplace=True)
            convrg_cond_genes_df = convrg_cond_genes_df.reindex(
                index=convrg_index_order[::-1]  # I think this reversing is specific to pandas barplotting
                ).fillna(0)
            
            ict_cond_key_genes = const.convrg_cond_gene_d[c]
            convrg_cond_genes_df['color'] = convrg_cond_genes_df.apply(lambda r: COND_KEY_GENE_COLOR if r.name in ict_cond_key_genes else 'white', axis=1)

            convrg_cond_genes_df['ALE experiment convergence'].plot(
                kind='barh', ax=axs[0], width=0.8, color=convrg_cond_genes_df['color'], edgecolor='black')
            axs[0].set_xlim([0, 1.1])
            axs[0].set_ylabel("")
            # axs[0].get_legend().remove()

            df = dva_cond_genes_of_interest_df.copy()  # Not applying any cutoffs in this NB

            sns.boxplot(
                ax=axs[1],
                data=df,
                x='log2FoldChange',
                y='target',
                order=convrg_index_order,
                flierprops = dict(marker='.', markeredgecolor='none',),
                boxprops=dict(facecolor='none'),
                width=0.8,
            )
            sns.stripplot(
                ax=axs[1],
                data=df,
                x='log2FoldChange',
                y='target',
                hue='variant source',
                hue_order=['WT alleleome', 'ICT ALE', 'ALEdb'],  # To match order with barplot.
                order=convrg_index_order,
                size=3,
                palette = const.VAR_SRC_COLOR_D,
                alpha=1,
                dodge=True,
            )
            axs[1].get_legend().remove()
            axs[1].yaxis.set_ticklabels([])
            axs[1].set_ylabel("")
            axs[1].yaxis.grid(True)
            axs[1].set_xlabel("")


            sns.boxplot(
                ax=axs[2],
                data=df,
                x='log2FoldChange',
                y='target',
                order=convrg_index_order,
                flierprops = dict(marker='.', markeredgecolor='none',),
                boxprops=dict(facecolor='none'),
            )
            sns.stripplot(
                ax=axs[2],
                data=df,
                x='log2FoldChange',
                y='target',
                hue='predicted variant effect',
                palette = var_eff_pal,
                order=convrg_index_order,
                hue_order=VAR_PRED_EFF_ORDER,
                size=3,
                alpha=1,
                dodge=True,
            )
            axs[2].get_legend().remove()
            axs[2].yaxis.set_ticklabels([])
            axs[2].set_ylabel("")
            axs[2].yaxis.grid(True)
            axs[2].set_xlabel("")

            subfigs[iy,ix].suptitle(c)

fig.supylabel('All key genes')
fig.supxlabel('ALE experiment convergence (left) and log2FoldChange colored by source (middle) and predicted variant effect (right)')
# fig.suptitle('Comparing ALE experiment key genes to significantly enriched Onyx designs')

from matplotlib.lines import Line2D

from matplotlib.patches import Patch

custom_patches = [
    Patch(facecolor='#FECB52', edgecolor='black'),
    Patch(facecolor='white', edgecolor='black')]
subfigs[iy,ix].legend(custom_patches, ['Condition specific key gene', 'General gene of interest'], loc='upper left', bbox_to_anchor=(0.1, 0.95))

# legend1_handles = [Line2D([0], [0], marker='o', color='w', label=var_src, markerfacecolor=color, markersize=10) for var_src, color in const.VAR_SRC_COLOR_D.items()]
# subfigs[iy,ix].legend(handles=legend1_handles, title='Variant Sources', loc='upper left', bbox_to_anchor=(0.1, 0.95))

# legend2_handles = [Line2D([0], [0], marker='o', color='w', label=var_eff, markerfacecolor=color, markersize=10) for var_eff, color in var_eff_pal.items()]
# subfigs[iy,ix].legend(handles=legend2_handles, title='Predicted Variant Effects', loc='upper left', bbox_to_anchor=(0.1, 0.75))

fig.text(0, 1, 'D', weight='bold', fontsize=14)

mpl.savefig("supp_f1d.svg")
```
